```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

advertising = pd.read_csv('/Users/EVC/GA/upstream/DAT_20_Students/EVCinNYC/Lecture07Classwork/Advertising.csv')
advertising = advertising.drop('Unnamed: 0', axis=1)

#show first 20 points of data set
#before was all with just advertising.head()
# ?? but not seeing this cut the number of points below
N = 200
advertising = advertising.head(N)

x = advertising['TV']
y = advertising['Sales']

plt.figure(figsize=(18,9))
axes = plt.gca()

axes.scatter(x, y, c = 'r', alpha=0.3)
# other paramaters in scatter
# s=## is plotsize
# alpha allows you to show the overlapping data better (going from 1 to 0.3)

y_mean = y.mean()

axes.set_title('Sales vs TV Ad Spend')
axes.set_xlabel('TV ad Spend')
axes.set_ylabel('Sales')

#adding in a horizontal line for the mean
axes.axhline(y_mean, linewidth=3, linestyle='--')
# can choose different linestyles

#Showing the variance among y -- so sum of the deviation
# x between mean of y and variable y
axes.vlines(x,y_mean,y, linestyle=':', alpha=0.3, colors='teal')
```

```python
#STEP TWO: Deriving beta_0 and beta_1 from set of points in x and y
from sklearn import linear_model
regressor = linear_model.LinearRegression()
#regressor.fit(x, y)  gave tuple index out of range
#have to create a new set of axis
regressor.fit(x[:,np.newaxis],y)

print regressor.coef_
print regressor.intercept_
```

```python
#underscores at the end mean they are "estimates" when looking at subsample of larger set 
reg_coef = regressor.coef_
reg_intercept = regressor.intercept_


x_min = x.min()
x_max = x.max()

plt.figure(figsize=(15,10))

axes_regression = plt.gca()
#axes.plot([x_min,x_max],[(x_min * reg_coef),(x_max * reg_coef)])
#use np.multiply when multiplying vectores because python does weird things
y_hat = np.multiply(regressor.coef_, x) + regressor.intercept_

#plot model with points fore ach value of x
axes_regression.scatter(x, y_hat, c='blue')
axes_regression.plot(x, y_hat, c='r')

axes_regression.grid(True)

axes_regression.vlines(x, y_hat, y, colors=['g'])
```
